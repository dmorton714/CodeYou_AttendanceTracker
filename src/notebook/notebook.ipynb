```python
def clean_staff_df(df: pd.DataFrame) -> pd.DataFrame:
    df.columns = df.columns.str.title()
    df = df.drop(['Alternate Name','Email','Region'],axis=1)
    df["Full Name"] = df['First Name'] + ' ' + df['Last Name']
    return df
    
```

```python
staff_list = clean_staff_df(staff_df)
staff_list
```

```python
def clean_mentor_df(df: pd.DataFrame) -> pd.DataFrame:
    df['Last Name'] = df['Last Name'].fillna('Unknown')
    df = df.drop(columns='Alternate Name')
    df["Full Name"] = df['First Name'] + ' ' + df['Last Name']
    return df
```

```python
mentor_list = clean_mentor_df(mentor_df)
mentor_list
```
